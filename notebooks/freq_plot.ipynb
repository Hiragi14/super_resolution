```python
from scipy.fft import dctn, idctn
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import torch
from torchvision.utils import save_image
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from numpy.fft import fft2, fftshift
```

```python
img = Image.open('./bird.jpg').convert('L')
```

```python
img_array = np.array(img)
```

```python
patch = img_array[500:600, 300:400]  # Set the top-left corner to zero
```

```python
def display_image(image_array, title=None):
    plt.imshow(image_array, cmap='gray')
    if title:
        plt.title(title)
    plt.axis('off')
    plt.savefig('patch_high.png', bbox_inches='tight', pad_inches=0.1)
    plt.show()
```

```python
display_image(patch)
```
